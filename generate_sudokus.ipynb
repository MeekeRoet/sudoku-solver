```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from matplotlib.font_manager import FontManager
from matplotlib.pyplot import imshow
from PIL import ImageFont

from sudoku_generator import SudokuGenerator

%matplotlib inline
```

```python
font_finder = FontManager()
font_path = font_finder.findfont('arial')
font = ImageFont.truetype(font_path, 35)
```

```python
data = []
gen = SudokuGenerator(dim=(500, 500), font=font, border=0.05)

for img_num in range(100):
    sudoku = gen.generate(np.random.uniform(0.05, 0.20))
    data.append(sudoku)
    
df = pd.DataFrame(data, columns=['img', 'corners'])
```

```python
plt.figure(figsize=(10, 10))
imshow(df.loc[0, 'img'])
plt.show()
```

```python
df.to_csv('data/sudokus.csv', index=False)
```
